```python
import urllib.request
from tqdm import tqdm
base_url = "http://www.centuryspring.com/extension/extension-springs.html?limit=36&showall=&p="

def download_page(num):
    url = base_url+str(num)

    response = urllib.request.urlopen(url)
    data = response.read()      # a `bytes` object
    text = data.decode('utf-8') # a `str`; this step can't be used if data is binary
    return text

pages = [download_page(i) for i in tqdm(range(1,122))]

    
```

```python
import pickle
with open('springs.p','wb') as f:
    pickle.dump(pages,f)
```

```python
from bs4 import BeautifulSoup

springs = []
for page in pages:
    soup = BeautifulSoup(page)
    table = soup.find('table', id = "category-products-table").tbody
    for row in table.findAll('tr'):
        values = [float(cell.text) for cell in row.findAll('td')[3:10]]
        springs.append(values+[row.findAll('td')[2].text])
```

```python
with open('springs.csv','w') as f:
    string_rows = [",".join([str(val) for val in row]) for row in springs]
    print(*string_rows, sep='\n', file=f)
```
